```python
import numpy as np
import matplotlib.pyplot as plt

# Configuracin del grfico
plt.figure(figsize=(10, 8))
plt.title('Comparativa de Curvas ROC segn el AUC', fontsize=16)
plt.xlabel('Tasa de Falsos Positivos (FPR)', fontsize=14)
plt.ylabel('Tasa de Verdaderos Positivos (TPR)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)

# Eje X base (FPR)
fpr = np.linspace(0, 1, 100)

# 1. AUC = 0.5 (Aleatorio - La Diagonal)
tpr_05 = fpr # La tasa de acierto es igual al error
plt.plot(fpr, tpr_05, linestyle='--', color='red', label='AUC = 0.5 (Aleatorio)', linewidth=2)

# 2. AUC = 0.7 (Modelo Regular)
# Simulamos una curva suave usando una funcin de potencia simple
tpr_07 = fpr**(1/2.5) 
plt.plot(fpr, tpr_07, color='orange', label='AUC  0.7 (Regular)', linewidth=2)

# 3. AUC = 0.9 (Modelo Excelente)
# Una potencia mayor crea una curva ms "cuadrada" hacia la esquina
tpr_09 = fpr**(1/8)
plt.plot(fpr, tpr_09, color='blue', label='AUC  0.9 (Excelente)', linewidth=2)

# 4. AUC = 1.0 (Perfecto)
# Esto es un ngulo recto: sube a 1 inmediatamente
plt.plot([0, 0, 1], [0, 1, 1], color='green', label='AUC = 1.0 (Perfecto)', linewidth=3)

# Mostrar leyenda y grfico
plt.legend(loc='lower right', fontsize=12)
plt.xlim([-0.02, 1.02])
plt.ylim([-0.02, 1.02])
plt.show()
```
