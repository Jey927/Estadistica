### **Variables Cuantitativas**

Variables Cuantitativas (o Num√©ricas): Se miden en una escala num√©rica. y en ese sentido existen dos tipos: 

1. **Discretas:** Solo pueden tomar valores enteros (contables). En el ejemplo de la tabla ser√≠a la **edad**
2. **Continuas:** Pueden tomar cualquier valor dentro de un rango (medibles).Ejemplo: **Altura, peso**.



### **Medidas de Tendencia Central***

1. **Media:** Es el promedio. Se calcula sumando todos los valores y dividiendo por el n√∫mero total de valores. 

2. **Mediana:** Es el valor central cuando los datos est√°n ordenados de menor a mayor. Deja el 50% de los datos por debajo y el 50% por encima. Es resistente a valores extremos.

3. **Moda:** Es el valor que aparece con mayor frecuencia en el conjunto de datos. Puede haber una moda, varias modas o ninguna.


En el caso de nuestra tabla podr√≠amos **sacar la media, mediana de la edad, la altura, el peso.** , por otro lado podriamos **Sacar la Moda de ubicaci√≥n, y el nivel de estudios**
